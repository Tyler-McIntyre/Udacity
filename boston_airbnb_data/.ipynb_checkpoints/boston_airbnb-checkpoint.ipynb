```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from decimal import Decimal, ROUND_HALF_UP, InvalidOperation

%matplotlib inline

df_listings = pd.read_csv('./data/listings.csv')
df_calendar = pd.read_csv('./data/calendar.csv')

print(df_listings.shape, df_calendar.shape)
```

```python
# First question...
# Are listings with high review scores rented out for more money and do they also have a higher percentage of satisfied guests?
```

```python
# Leave the original data frames intact
listings_copy = df_listings.copy() # Avoid the SettingWithCopyWarning
listings_copy.head()
```

```python
# Leave the original data frames intact
calendar_copy = df_calendar.copy() # Avoid the SettingWithCopyWarning
calendar_copy
```

```python
# Data prep
```

```python
# Drop columns with all missing values
print(listings_copy.shape, calendar_copy.shape) # Before
listings_copy.dropna(axis=1, how='all')
calendar_copy.dropna(axis=1, how='all')
print(listings_copy.shape, calendar_copy.shape) # After
```

```python
# Drop columns with missing review_scores_rating
print(listings_copy.shape) # Before
listings_copy = listings_copy.dropna(subset=['review_scores_rating'])
print(listings_copy.shape) # After
```

```python
# Pull the number_of_reviews, and review_scores_rating from listings
listings_copy = listings_copy[['id', 'number_of_reviews','review_scores_rating', 'property_type', 'room_type', 'bathrooms', 'bedrooms', 'beds', 'cleaning_fee']]
listings_copy.head()
```

```python
# Pull the listing_id and price from calendar_copy (calendar.csv)
calendar_copy = calendar_copy[['listing_id','price']]
calendar_copy.head()
```

```python
# The price column is an object. We'll need to convert that to decimal to get the mean later
object_cols = calendar_copy.select_dtypes(include=['object']).any()
object_cols
```

```python
def string_to_decimal(s):    
    # Remove dollar signs and commas, handle NaN and empty strings
    if pd.isna(s) or s in ['', 'NaN', 'nan']:
        return np.nan  # Return NaN values we'll drop those later
    try:
        # Remove currency symbols and commas, then convert to Decimal
        cleaned_value = s.replace('$', '').replace(',', '')
        return Decimal(cleaned_value).quantize(Decimal('0.01'), rounding=ROUND_HALF_UP)
    except InvalidOperation:
        return np.nan  # Return NaN invalid conversion cases

# Convert to decimal
calendar_copy.loc[:, 'price'] = calendar_copy['price'].apply(string_to_decimal) # Avoid the SettingWithCopyWarning with .loc value
```

```python
# Drop NaN rows
calendar_copy = calendar_copy.dropna(subset=['price'])
pd.concat([calendar_copy.head(), calendar_copy.tail()])
```

```python
# Group by and average the price per listing id
print(calendar_copy.shape) # Before
calendar_copy['avg_listed_price'] = calendar_copy.groupby('listing_id')['price'].transform('mean')
calendar_copy = calendar_copy.drop_duplicates(subset='listing_id', keep='first')
calendar_copy.shape # After, leaves 2906 unique listing_id's
```

```python
# Rename the id column to match the calendar id column
listings_copy = listings_copy.rename(columns={'id': 'listing_id'})
listings_copy[:1]
```

```python
# Merge the listings_df on the calendar_df
merged_df = pd.merge(listings_copy, calendar_copy, on='listing_id', how='inner')
merged_df.head()
```

```python
# Plot the Review Ratings vs Avg Listed Price
plt.scatter(merged_df['review_scores_rating'], merged_df['avg_listed_price'])
plt.xlabel('Reviews Score Rating')
plt.ylabel('Avg Listed Price')
plt.title('Review Rating vs. Avg Listed Price')
plt.show()
```

```python
# The previous frame showed we had some drastic outliers that could be cleaned up
print(merged_df.shape)

# Function to remove outliers for each column
def remove_outliers(df, columns):
    indices_to_drop = set()
    # iterate over the numerical columns
    for col in columns:
        # Remove using the Interquartile Range (IQR) Method (removes the 25th and 75th percentile outliers)
        Q1 = df[col].quantile(0.15)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Find indices above and below the lower/upper bounds
        outliers_indices = df[(df[col] < lower_bound) | (df[col] > upper_bound)].index
        indices_to_drop.update(outliers_indices)
        print(sorted(indices_to_drop))
        # Drop the rows that contain outliers
        df_cleaned = df.drop(index=indices_to_drop)
        return df_cleaned

# Apply the func
cleaned_df = remove_outliers(merged_df, ['review_scores_rating', 'avg_listed_price'])
```

```python
cleaned_df.sort_values(by='avg_listed_price', ascending=False)
```

```python
# Re-plot Review Ratings vs Avg Listed Price
plt.scatter(cleaned_df['review_scores_rating'], cleaned_df['avg_listed_price'])
plt.xlabel('Review Score Rating')
plt.ylabel('Avg Listed Price')
plt.title('Review Rating vs. Avg Listed Price')
plt.show()
```

```python
sorted_df = cleaned_df.sort_values(by='avg_listed_price', ascending=False)
```

```python
# Spot-clean the top value
df_spot_cleaned = sorted_df[1:]

df_spot_cleaned.head()
```

```python
# Re-plot Review Ratings vs Avg Listed Price
plt.scatter(df_spot_cleaned['review_scores_rating'], df_spot_cleaned['avg_listed_price'])
plt.xlabel('Review Score Rating')
plt.ylabel('Avg Listed Price')
plt.title('Review Rating vs. Avg Listed Price')
plt.show()
```

```python
# Bin the data
x_bins = np.linspace(df_spot_cleaned['review_scores_rating'].min(), df_spot_cleaned['review_scores_rating'].max(), 20)
y_bins = np.linspace(df_spot_cleaned['avg_listed_price'].min(), df_spot_cleaned['avg_listed_price'].max(), 20) 

# Calculate the 2D histogram
h, xedges, yedges = np.histogram2d(df_spot_cleaned['review_scores_rating'], df_spot_cleaned['avg_listed_price'], bins=[x_bins, y_bins])

# Create the heatmap
fig, ax = plt.subplots()
cax = ax.imshow(h.T, extent=[xedges[0], xedges[-1], yedges[0], yedges[-1]], origin='lower', cmap='YlGnBu', aspect='auto')

# Add colorbar
cbar = fig.colorbar(cax)
cbar.ax.set_ylabel('Counts')

# Labels
ax.set_xlabel('Review Score Rating')
ax.set_ylabel('Avg Listed Price')

plt.show()
```

```python
# Conclusion for question 2:
# High Price and High Review Score: There is a significant concentration (dark blue area) at the highest review score of 100, 
# which corresponds to the highest price range observed on the heatmap. This suggests a correlation where items with perfect
# review scores also tend to be listed at higher prices.

# The concentration of high review scores suggests that customers are largely satisfied with their rental
```

```python
# Question 2
# What accommodation types & sizes frequently receive the highest reviews?
```

```python
# Remove listings with a low amount of reviews.
# Reason: Listings with a higher amount of reviews should have more reliable
print('avg review count:', df_spot_cleaned['number_of_reviews'].mean()) # avg amount of reviews is 29 ish
df_dropped_low_reviews = df_spot_cleaned[df_spot_cleaned['number_of_reviews'] < 10].copy()
```

```python
# find the average score for each accommodation type
room_type_score = df_dropped_low_reviews.groupby(by='room_type')['review_scores_rating'].transform('mean')

df_dropped_low_reviews.loc[:,'room_type_score'] = room_type_score
set(sorted(room_type_score))
```

```python
df_room_types = df_dropped_low_reviews.drop_duplicates('room_type_score')
df_room_scores_sorted = df_room_types.sort_values(by='room_type_score')

plt.plot(df_room_scores_sorted['room_type'], df_room_scores_sorted['room_type_score'])
plt.scatter(df_room_scores_sorted['room_type'], df_room_scores_sorted['room_type_score'], color='blue', zorder=5)

plt.xlabel('Room Types')
plt.ylabel('Avg Review Score')

plt.title('Room Types Vs Avg Review Score')
plt.show()
```

```python
# This suggests that, within this dataset, shared rooms tend to have higher average review scores compared to private rooms and entire homes/apartments.

# The differences in average review scores between the room types are relatively small, with all scores falling between approximately 92 and 94 percent. 
# This indicates that while there is a trend, the variation in review scores among different room types is not very large.

# Overall this suggests that the type of room does not have a clear impact or relationship to higher review scores
```

```python
# Question 3
# Do accommodations that charge a higher-than-average cleaning fee receive a lower review score?
```

```python
# Drop the rows that don't have a cleaning fee listed
df_dropped_low_reviews = df_dropped_low_reviews.dropna(subset=['cleaning_fee'], how='all')
df_dropped_low_reviews['cleaning_fee'].isna().any()
```

```python
# Compare the average cleaning fee against the average review score
# If the listing is charging above the average does it correlate with a low review score?

# I want to see the relationship between the cleaning fee and review score, and what the listing is charging
# Above or below the average cleaning fee I want to compare that to the review score
```

```python
# Convert the cleaning_fee to decimal values
df_dropped_low_reviews.loc[:, 'cleaning_fee'] = df_dropped_low_reviews['cleaning_fee'].apply(string_to_decimal)
```

```python
# Calculate the average review score and average cleaning fee
average_review_score = df_dropped_low_reviews['review_scores_rating'].mean()
average_cleaning_fee = df_dropped_low_reviews['cleaning_fee'].mean()

# Create a new DataFrame to hold the relative differences
comparison_df = pd.DataFrame()
comparison_df['review_score'] = df_dropped_low_reviews['review_scores_rating']
comparison_df['cleaning_fee_relative'] = df_dropped_low_reviews['cleaning_fee'] - Decimal(average_cleaning_fee)

# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(comparison_df['cleaning_fee_relative'], comparison_df['review_score'], alpha=0.6)

# Add a horizontal line at the average review score
plt.axhline(y=average_review_score, color='r', linestyle='--', label='Average Review Score')

# Add a vertical line at the point where the cleaning fee equals the average
plt.axvline(x=0, color='g', linestyle='--', label='Average Cleaning Fee')

# Label the axes
plt.xlabel('Cleaning Fee Relative to Average')
plt.ylabel('Review Score')
plt.title('Comparison of Review Scores and Cleaning Fees Relative to Averages')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

```python
# Conclusion for question 3:
# High Cleaning Fee Influence: Listings with significantly higher cleaning fees (to the right of the green dashed line) 
# do not show a consistent trend of having lower review scores. 
# There are listings with both high and low review scores at higher cleaning fees.
```
